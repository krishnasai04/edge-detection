```
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread('test7.jpg', cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded successfully
if image is None:
    print("Error: Could not open the image.")
    exit()

```

```
plt.figure(figsize=(15, 15))
plt.subplot(4, 3, 10), plt.imshow(image, cmap='gray'), plt.title('Original Image')

plt.show()
```

```
# Apply Sobel operator
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

# Apply Prewitt operator
kernel_prewitt_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
kernel_prewitt_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
prewitt_x = cv2.filter2D(image, cv2.CV_64F, kernel_prewitt_x)
prewitt_y = cv2.filter2D(image, cv2.CV_64F, kernel_prewitt_y)
prewitt_magnitude = np.sqrt(prewitt_x**2 + prewitt_y**2)

# Apply Roberts operator
kernel_roberts_x = np.array([[1, 0], [0, -1]])
kernel_roberts_y = np.array([[0, 1], [-1, 0]])
roberts_x = cv2.filter2D(image, cv2.CV_64F, kernel_roberts_x)
roberts_y = cv2.filter2D(image, cv2.CV_64F, kernel_roberts_y)
roberts_magnitude = np.sqrt(roberts_x**2 + roberts_y**2)
```

```
# Create a single window with multiple subplots
plt.figure(figsize=(15, 15))

plt.subplot(4, 3, 1), plt.imshow(np.abs(sobel_x), cmap='gray'), plt.title('Sobel X')
plt.subplot(4, 3, 2), plt.imshow(np.abs(sobel_y), cmap='gray'), plt.title('Sobel Y')
plt.subplot(4, 3, 3), plt.imshow(np.uint8(sobel_magnitude), cmap='gray'), plt.title('Sobel Magnitude')

plt.subplot(4, 3, 4), plt.imshow(np.abs(prewitt_x), cmap='gray'), plt.title('Prewitt X')
plt.subplot(4, 3, 5), plt.imshow(np.abs(prewitt_y), cmap='gray'), plt.title('Prewitt Y')
plt.subplot(4, 3, 6), plt.imshow(np.uint8(prewitt_magnitude), cmap='gray'), plt.title('Prewitt Magnitude')

plt.subplot(4, 3, 7), plt.imshow(np.abs(roberts_x), cmap='gray'), plt.title('Roberts X')
plt.subplot(4, 3, 8), plt.imshow(np.abs(roberts_y), cmap='gray'), plt.title('Roberts Y')
plt.subplot(4, 3, 9), plt.imshow(np.uint8(roberts_magnitude), cmap='gray'), plt.title('Roberts Magnitude')
plt.show()
```
